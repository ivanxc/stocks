```python
import pandas as pd
import json
            
filename = "iRobot Corp.json"
srings = []
with open(filename) as f:
    strings = f.readlines()
    
with open("new_" + filename, "w") as f:
    f.write("[")
    for string in strings:
        if string.find("time") == -1:
            continue
        else:
            f.write(string.replace("\n", ",\n"))
    f.write("]")    
    
df = pd.read_json("new_" + filename)
for i in range(len(df)):
    df.iloc[i, 6] = len(df.iloc[i, 6])
    df.iloc[i, 7] = len(df.iloc[i, 7])
df.drop_duplicates(subset="time", keep = 'first', inplace = True)
del df['face_value']
del df['trade_status']
df
```
